```python
import turtle
import time

#Creating the Screen for display

screen=turtle.Screen()
screen.setup(width=800, height=800)
screen.bgcolor("black")
screen.title("Analog Clock")
screen.tracer(0)        #To avoid animation

#Creating a tool to Draw
stylus= turtle.Turtle()
stylus.speed(0)
stylus.pensize(4)

#Creating the outline of the clock
def outline(stylus):
	stylus.up()
	stylus.goto(0,350)
	stylus.setheading(180)
	stylus.color("Red")
	stylus.pendown()
	stylus.circle(350)

#Creating the hours
def hours_outline(stylus):
	stylus.color("Brown")
	stylus.up()
	stylus.goto(0,0)
	stylus.setheading(90)
	stylus.rt(30)

	hr=0
	for i in range (0,12):
		hr+=1
		stylus.fd(250)
		stylus.write(str(hr), align="center", font=("oswald",20, "normal"))
		stylus.goto(0,0)
		stylus.rt(30)

#Creating minutes
def minutes_outline(stylus):
	stylus.color("Red")
	stylus.up()
	stylus.goto(0,0)
	stylus.setheading(90)
	i=0
	while(i<=360):
		stylus.rt(i)
		if(i%30!=0):
			stylus.pensize(1)
			stylus.fd(300)
			stylus.down()
			stylus.fd(50)
			stylus.up()
			stylus.goto(0,0)
		else:
			stylus.pensize(5)
			stylus.fd(250)
			stylus.down()
			stylus.fd(100)
			stylus.up()
			stylus.goto(0,0)	
		stylus.setheading(90)
		i+=6

#Creating the hour hand
def hour_hand(stylus,h):
	stylus.up()
	stylus.goto(0,0)
	stylus.down()
	stylus.setheading(90)
	stylus.rt(h*30)
	stylus.fd(100)

#Creating the minute hand
def minute_hand(stylus,m):
	stylus.up()
	stylus.goto(0,0)
	stylus.down()
	stylus.setheading(90)
	stylus.rt(m*6)
	stylus.fd(200)

#Creating the second hand
def second_hand(stylus,s):
	stylus.up()
	stylus.goto(0,0)
	stylus.down()
	stylus.setheading(90)
	stylus.rt(s*6)
	stylus.fd(225)



while True:
	h= int(time.strftime("%I"))
	m= int(time.strftime("%M"))
	s= int(time.strftime("%S"))
	outline(stylus)
	hours_outline(stylus)
	minutes_outline(stylus)
	hour_hand(stylus,h)
	minute_hand(stylus,m)
	second_hand(stylus,s)
	screen.update()     # To see the animation
	time.sleep(1)       # To avoid blinking and show animation after 1
	stylus.clear()      # To avoid overlapping
screen.mainloop()
```
